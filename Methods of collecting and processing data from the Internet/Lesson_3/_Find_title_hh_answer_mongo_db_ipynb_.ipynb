**Функцию, которая будет добавлять только новые вакансии в вашу базу.**

Решение - новой вакансией будет считаться только та, которая будет иметь новую ссылку. Другие поля в плане уникальности не помогут.

```python
def add_new_job(job):
  link = job['Link']
  if len(list(collection.find({'Link': link})))>0:
    print('NOT NEW JOB')
  else:
    collection.insert_one(job)
```

```python
#протестирую
tmp = {'Currency': 'руб. ',
  'Link': 'https://hh.ru/vacancy/49629106?from=vacancy_search_list&_TEST_query=%D1%8D%D0%BA%D0%BE%D0%BB%D0%BE%D0%B3',
  'Max_salary': 900000,
  'Min_salry': None,
  'Name site': 'https://hh.ru',
  'Title': 'Специалист по охране труда/эколог - ПРОВЕРКА'}

add_new_job(tmp)
```
