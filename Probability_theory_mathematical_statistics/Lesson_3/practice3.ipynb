Для точечного оценивания параметров случайной величины используются различные статистики. __Статистика__ — это любая функция от выборки.

Познакомимся со статистиками на примере датасета с данными по хоккеистам. Данные доступны [здесь](https://habr.com/post/301340/).

Импортируем данные.

```python
import pandas as pd
```

```python
df = pd.read_csv('hockey_players.csv')
df.head()
```

Пусть дана выборка $X = (x_1, x_2, \dots, x_n)$ из значений одномерной случайной величины. Одной из наиболее естественных статистик таких выборок является __среднее арифметическое__ (или __выборочное среднее__). Оно обозначается как $\overline{X}$:
$$\overline{X} = \dfrac{1}{n} \displaystyle\sum_{i = 1}^n x_i$$

Выборочное среднее является __оценкой__ для математического ожидания. Это означает, что, как правило, чем больше элементов в выборке, тем ближе выборочное среднее этой выборки к математическому ожиданию соответствующей случайной величины.

__Пример 1__

Найдем выборочное среднее роста хоккеистов из датасета по формуле.

```python
mean_height = df['height'].sum() / df['height'].shape[0]
mean_height
```

Эту же величину можно вычислить сразу с помощью метода `.mean`.

```python
df['height'].mean()
```

__Выборочная дисперсия__, как следует из названия, оценивает дисперсию случайной величины:
$$\sigma_X^2 = \dfrac{1}{n - 1} \displaystyle\sum_{i = 1}^n \left( x_i - \overline{X} \right)^2$$

Такая формула для оценки дисперсии называется __несмещённой__ за счёт этой единицы в знаменателе. Существует также __смещённая__ оценка дисперсии:
$$\sigma_{X, \: biased}^2 = \dfrac{1}{n} \displaystyle\sum_{i = 1}^n \left( x_i - \overline{X} \right)^2$$

Однако, она в практических задачах не используется.

Почему смещённые оценки — это плохо, а также почему вычитают именно $1$ — читайте в дополнительных материалах к уроку.

__Пример 2__

Посчитаем выборочную дисперсию роста хоккеистов.

```python
df.shape
```

```python
((df['height'] - df['height'].mean()) ** 2).sum() / (df['height'].shape[0] - 1)
```

Массив `(df['height'] - df['height'].mean()) ** 2` представляет собой элементы из `df['height']`, из которых поэлементно вычли их среднее, а затем возвели в квадрат:

```python
(df['height'] - df['height'].mean()) ** 2
```

То же с помощью готовой функции `.var`:

```python
df['height'].var(ddof=1)
```

Использованный параметр `ddof` (_Delta Degrees of Freedom_, или _Разность степеней свободы_) контролирует то, сколько мы вычитаем из $n$ при вычислении оценки дисперсии. Для несмещённой оценки используется `ddof=1`. Подставив `ddof=0`, можно посчитать смещённую оценку:

```python
df['height'].var(ddof=0)
```

```python
((df['height'] - df['height'].mean()) ** 2).sum() / df['height'].shape[0]
```

_Замечание_: с этим параметром стоит быть осторожным, потому что его значение по умолчанию отличается в библиотеках `numpy` и `pandas`:

```python
help(pd.DataFrame.var)
```

```python
import numpy as np
```

```python
help(np.var)
```

Вообще, дисперсия является не очень наглядной мерой разброса, поскольку имеет другой масштаб. Поэтому часто наряду с дисперсией используют __среднее квадратическое отклонение__, равное корню из дисперсии.

Оценивается среднее квадратическое отклонение аналогично дисперсии:
$$\sigma_X = \sqrt{\dfrac{1}{n - 1} \displaystyle\sum_{i = 1}^n \left( x_i - \overline{X} \right)^2}$$

__Пример 5__

Среднее квадратическое отклонение роста хоккеистов:

```python
np.sqrt(((df['height'] - df['height'].mean()) ** 2).sum() / (df['height'].shape[0] - 1))
```

То же с помощью готовой функции `.std`:

```python
df['height'].std(ddof=1)
```

Итак, мы видим, что в среднем рост хоккеистов отклоняется от своего среднего примерно на $5.4676$.

### Медиана, квантили

__Медиана__ — такое значение $t$, что половина элементов из выборки меньше, либо равна $t$, и, соответственно, половина больше, либо равна $t$.

Медиана представляет собой _середину_ выборки: если отсортировать элементы выборки по возрастанию, то медиана приходится на середину. 

Медиана может приходиться как на промежуток между элементами выборки, так и на конкретный элемент.

__Пример 6__

Посчитаем медиану роста хоккеистов. Для начала воспользуемся определением. Нужно отсортировать значения из выборки и взять середину этого массива. Посчитаем размер выборки:

```python
height = sorted(df['height'])

length = len(height)
length
```

Правило вычисления медианы следующее:
* Если выборка имеет чётный размер, то медиана будет находиться где-то посередине между крайним правым элементом левой половины и крайним левым элементом правой половины.
* Если выборка имеет нечётный размер, то медианой будет элемент, стоящий на позиции $[n / 2] + 1$, где $[x]$ — целая часть от числа $x$.

В нашем случае выборка имеет чётный размер. Как найти индекс крайнего правого элемента левой половины?

В левой половине $n / 2$ элементов, значит, крайний правый из них будет стоять на позиции $n / 2$. В нашем случае это индекс `length / 2 - 1`, поскольку нумерация начинается с нуля. Соответственно, следующий за ним элемент будет иметь индекс `length / 2`.

```python
height[length // 2 - 1], height[length // 2]
```

В нашем случае оба значения равны. Поскольку медиана должна находиться между ними, она тоже равна $184$. 

```python
median = (height[length // 2 - 1] + height[length // 2]) / 2
median
```

Посмотрим, сколько процентов выборки лежат слева и справа от медианы.

```python
(df['height'] <= median).sum() / length, (df['height'] >= median).sum() / length
```

```python
df['height'].median()
```

Медиана также является мерой среднего, как выборочное среднее и мода.

```python
df['height'].mean()
```

В случаях, когда число элементов в выборке нечётно, медианой является значение, которое стоит ровно посередине отсортированного массива. Если же в массиве чётное число значений, ровную середину посчитать невозможно, поэтому медианой может служить любое значение между крайним правым значением левой половины и крайним левым значением правой половины.

```python
help(np.quantile)
```

### Квантили

Медиана является частным случаем более общего понятия — _квантиля_. 

Пусть $\alpha \in (0, 1)$. __Квантиль порядка $\alpha$__ — такое число $t_\alpha$, что «$\alpha$ процентов» всех элементов выборки меньше $t_\alpha$ и, соответственно, «$(1 - \alpha)$ процентов» элементов — больше $t_\alpha$.

Как и в случае с медианой, квантиль может как приходиться на один из элементов выборки, так и лежать где-то между ними.

Из определения следует, что медиана является квантилем порядка $0.5$. Кроме того, часто используют:

* __первый квартиль__ — квантиль порядка $0.25$ (т.е. значение, которое не превышают $25 \%$ значений из выборки),
* __второй квартиль__ — то же, что и медиана,
* __третий квартиль__ — квантиль порядка $0.75$.

Также могут встречаться:
* __децили__ — то же, что и квартили, но делим мы не на $4$ части, а на $10$. Например, медиана будет пятым децилем,
* __перцентили__ — это просто другой способ задать квантиль. Здесь мы используем не долю $\alpha \in (0, 1)$, а процент. Например, третий квартиль будет $75$-перцентилем.

__Пример 7__

Посчитаем первый и третий квартили значений роста хоккеистов. Вычисления аналогичны случаю вычисления медианы, однако нам нужно делить массив не в пропорции $1 : 1$, а в пропорции $\alpha : (1 - \alpha)$.

Посчитаем первый квартиль, т.е. квартиль порядка $0.25$. Процесс похож на вычисление медианы, просто делить будем не на $2$, а на $4$. Правило вычисления квартиля:
* Если размер массива делится на 4 нацело, то первый квартиль будет находиться между крайним правым элементом из левой четверти (в отсортированном массиве) и элементом, следующим за ним.
* Если размер массива на 4 не делится, то первым квартилем будет элемент, стоящий на позиции $[n / 4] + 1$, где $[x]$ — целая часть числа $x$.

Итак, остаток от деления длины массива на $4$:

```python
length % 4
```

На $4$ не делится. Значит, квартиль придётся на конкретный элемент, а именно элемент с индексом `length // 4` (опять же поскольку индексация происходит с нуля).

```python
q1 = height[length // 4]
q1
```

```python
(df['height'] <= q1).sum() / length, (df['height'] >= q1).sum() / length
```

```python
df['height'].quantile(0.25)
```

Третий квартиль вычисляется аналогично, но будет стоять на позиции `3 * length // 4`:

```python
q3 = height[3 * length // 4]
q3
```

```python
(df['height'] <= q3).sum() / length, (df['height'] >= q3).sum() / length
```

```python
df['height'].quantile(0.75)
```

В этот метод можно подать список значений, тогда посчитаются различные квантили:

```python
df['height'].quantile([0.25, 0.5, 0.75])
```

```python
df['height'].describe()
```

_Замечание_. Предложенный выше метод вычисления квантилей может отличаться от других имплементаций (в том числе `numpy`). Я не знаю, почему они считают квантили так, как считают, поэтому они получаются неправильные. Например, посчитаем для массива ниже первый квартиль нашим способом и с помощью `numpy`:

```python
a = np.linspace(0, 1000, 102)

q_our = sorted(a)[a.shape[0] // 4]  # поскольку размер массива не делится на 4
q_numpy = np.quantile(a, 0.25)

q_our, q_numpy
```

Проверим процент значений слева и справа от этих квартилей:

```python
(a <= q_our).sum() / a.shape[0], (a >= q_our).sum() / a.shape[0]
```

```python
(a <= q_numpy).sum() / a.shape[0], (a >= q_numpy).sum() / a.shape[0]
```

Конечно, разница не велика, так что в практических задачах это вряд ли сыграет важную роль. Но разница есть, и её стоит отметить.

__Интерквартильный размах__ — это отрезок между первым и третьим квартилями. Это отрезок, в который попадают $50 \%$ значений выборки.

Интерквартильный размах используется для измерения разброса значений выборки вокруг среднего. Иногда его использование оказывается более предпочтительным, чем использование среднего квадратического отклонения, поскольку не учитывает выбросы в данных.

Например, рассмотрим синтетическую выборку, в которой $999$ раз встречается значение $1$, и всего однажды появляется значение $10000$.

```python
ar = np.ones(1000)
ar[0] = 10000

ar.mean(), ar.std()
```

При этом, интерквартильный размах:

```python
np.quantile(ar, [0.25, 0.75])
```

#### Квантиль случайной величины

Понятие квантиля также можно определить для случайной величины. Суть определения такая же, что и в случае выборки, но выглядит немного страшнее.

__Квантилем порядка $\alpha$ случайной величины $X$__ называется такое значение $t_\alpha$, что
$$P(X \leq t_\alpha) = \alpha, \:\: P(X \geq t_\alpha) = 1 - \alpha$$

Идея та же: в доле $\alpha$ всех случаев значение случайной величины $X$ окажется меньше $t_\alpha$ и в доле $(1 - \alpha)$ случаев — больше $t_\alpha$.

Теоретические квантили постоянно используются в статистических методах (о таких мы поговорим на уроке 5). Они позволяют решить следующую задачу: имеется случайная величина $X$ и значение вероятности $\alpha \in (0, 1)$. Требуется найти пороговое значение $t$, такое, что $P(X \leq t) = \alpha$.

Это и есть квантиль порядка $\alpha$.

### Графическое представление данных

Для визуализации распределения значений выборки часто используется гистограмма. Как строится гистограмма?

1. По оси $x$ откладываются всевозможные значения из выборки. 
2. Вся ось разбивается на какое-то заданное число одинаковых отрезков.
3. Для каждого отрезка вычисляется число значений выборки, которые лежат в этом отрезке, и это число откладывается по оси $y$.

__Пример 8__

Построим гистограмму значений роста. Для этого можно использовать метод `.hist`. В качестве аргумента `bins` сюда подаётся то число одинаковых отрезков, на которое мы разбиваем ось $x$.

```python
from matplotlib import pyplot as plt

plt.style.use('seaborn-whitegrid')
%config InlineBackend.figure_formats = ['svg']
```

```python
df['height'].hist()
```

Построим гистограмму ещё раз, указав в качестве аргумента `bins` число $20$, чтобы получить более детальную картину:

```python
df['height'].hist(bins=20)
```

```python
df['height'].hist(bins=50)
```

```python
df['height'].hist(bins=5)
```

Гистограмма по форме напоминает график распределения вероятностей случайной величины. Нужно лишь нормировать откладываемые по оси $y$ значения, чтобы сумма площадей колонок стала равной $1$. Это можно сделать с помощью аргумента `density`.

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4))

df['height'].hist(ax=ax1)
df['height'].hist(ax=ax2, density=True)
```

Другой способ визуализировать одномерные данные — __boxplot__ или __ящик с усами__. В самом ящике отмечены квартили $Q_1$, $Q_2$ (медиана), $Q_3$. «Усы» здесь — границы отрезка
$$\left[ Q_1 - 1.5 \times IQR, \: Q_3 + 1.5 \times IQR \right],$$
где $IQR$ — интерквартильное расстояние.

__Пример 9__

Построим `boxplot` роста хоккеистов.

```python
df[['height']].boxplot()
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=df, y='height')
```

Всё, что выходит за границы этого отрезка, считается выбросами (отмечены кружками). Посчитаем, сколько значений роста оказалось за пределами «усов».

```python
q1, q3 = df['height'].quantile([0.25, 0.75])

iqr = q3 - q1

boxplot_range = (q1 - 1.5 * iqr, q3 + 1.5 * iqr)
boxplot_range
```

```python
outliers = df.loc[(df['height'] < boxplot_range[0]) | (df['height'] > boxplot_range[1])]

outliers.shape[0]
```

```python
outliers
```

Доля выбросов:

```python
outliers.shape[0] / df.shape[0]
```
